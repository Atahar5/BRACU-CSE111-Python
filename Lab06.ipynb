```python
#Q/A No - 1
class Student:
    id = 0
    def __init__(self,name,department,age,cgpa):
        Student.id = Student.id + 1
        self.name = name
        self.dept = department
        self.age = age
        self.cgpa = cgpa   
    def get_details(self):
        print(f"ID: {Student.id}")
        print(f"Name: {self.name}")
        print(f"Department: {self.dept}")
        print(f"Age: {self.age}")
        print(f"CGPA: {self.cgpa}")
        
    @classmethod
    def from_String(cls,data):
        name,dept,age,cgpa = data.split("-")
        return cls(name,dept,age,cgpa)
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print("\n'Class variables' are shared across all objects while 'instance variables' are for data unique to each instance.")
print("\nAn 'instance method' uses the information contained in the instance. But A 'class method' uses information about the class")
```

```python
#Q/A No - 2
class Dates:
    def __init__(self, date):
        self.date = date
        
    @classmethod
    def toDashDate(self, dashDate):
        return self(dashDate.replace('/', '-'))
    def getDate(self):
        return self.date
date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
    print("Equal")
else:
    print("Unequal")
print("date1 is '05-09-2020' and date2 was '05-09-2020'.\nIn the conditional statement it prints “Equal” because “date1” and “date2” contains same date.")
```

```python
#Q/A No - 5
class Fruit:
  fruitCount = 0
  def __init__(self, name, count):
    self.name = name
    self.count = count
    Fruit.fruitCount += count

  @staticmethod
  def saySomethingGood():
    print("Fruits are good for health")

  @staticmethod
  def resetcount():
    Fruit.fruitCount = 0

apples = Fruit("Apple", 3);
pears = Fruit("Pear", 4);
print(apples.name, apples.count)
print(pears.name,pears.count)
print("Total number of fruits", Fruit.fruitCount)
Fruit.saySomethingGood()
Fruit.resetcount()
print("Total number of fruits", Fruit.fruitCount)
```

```python
#Q/A No - 6
class Cat:
    Number_of_cats=0
    def __init__(self,x,y):
        self.color=x
        self.work=y
        Cat.Number_of_cats=Cat.Number_of_cats+1
        
    @classmethod
    def no_parameter(cls):
        return cls("White","sitting")
    
    @classmethod
    def first_parameter(cls,x):
        return cls(x,"sitting")
    
    @classmethod
    def second_parameter(cls,y):
        return cls("Grey",y)
    def changeColor(self,x):
        self.color=x
    def printCat(self):
        print(self.color," cat is ",self.work)
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Q/A No - 7
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, r, h):
        self.radius = r
        self.height = h
        print('Default radius={} and height={}.'.format(Cylinder.radius, Cylinder.height))
        Cylinder.radius = r
        Cylinder.height = h
        print('Updated: radius={} and height={}.'.format(r, h))
        
    @classmethod
    def swap(self, h, r):
        return self(r, h)
    
    @classmethod
    def changeFormat(self, s):
        s_arr = s.split('-')
        return self(float(s_arr[0]), float(s_arr[1]))
    
    @staticmethod
    def area(r, h):
        res = 2 * math.pi * r * r + 2 * math.pi * r * h
        print('Area:', res);
        
    @staticmethod
    def volume(r, h):
        res = math.pi * r * r * h
        print('Volume:', res)
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```
